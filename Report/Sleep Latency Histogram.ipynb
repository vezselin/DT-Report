```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Convert time strings like "0:30:00" into minutes.
def clean_latency_time_string(time_str):
    if pd.isna(time_str):
        return np.nan
    try:
        parts = str(time_str).split(':')
        # We only care about hours and minutes
        hours = int(parts[0])
        minutes = int(parts[1])
        return (hours * 60) + minutes
    except:
        return np.nan

try:
    #Load the file
    file_path = 'survey-responses.csv'
    df = pd.read_csv(file_path)

    df['Sleep Latency (Minutes)'] = df['How long did it take you to fall asleep?'].apply(clean_latency_time_string)

    #Histogram
    plt.figure(figsize=(10, 6))
    plt.hist(df['Sleep Latency (Minutes)'].dropna(), bins=10, edgecolor='black', alpha=0.7)
    plt.title('Distribution of Time to Fall Asleep (Sleep Latency)')
    plt.xlabel('Sleep Latency (Minutes)')
    plt.ylabel('Number of Nights (Frequency)')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show() # Display the plot

except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```
